```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
cars_df=pd.read_csv("/content/mtcars (1).csv")
print(cars_df)
a=plt.scatter(cars_df['mpg'],cars_df['wt'],marker='*')
print(a)
plt.xlabel("MPG")
plt.ylabel("Weight")
plt.title('relationship between the weight of the car and MPG')
plt.show()
trans=cars_df['am'].value_counts()
print(trans)
trans.plot.bar()
plt.xlabel("Transmission")
plt.title('Bar plot to check the Frequency distribtion of trasmission type of cars')
plt.show()
cars_df['mpg'].plot(kind='box')
plt.xlabel('MPG')
plt.title('Box and whisker Plot of MPG')
plt.show()
min_value=np.min(cars_df['mpg'])
q1=np.percentile(cars_df['mpg'],25)
median=np.median(cars_df['mpg'])
q3=np.percentile(cars_df['mpg'],75)
max_value=np.max(cars_df['mpg'])
print('minimumValue:',min_value)
print('Q1:',q1)
print('Median:',median)
print('Q3:',q3)
print('maximum Value:',max_value)
```
